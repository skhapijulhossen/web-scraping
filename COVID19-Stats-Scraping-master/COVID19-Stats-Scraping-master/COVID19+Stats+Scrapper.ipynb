# Scrapping of COVID-19 Stats of World

```python
# import Necessary Module
from urllib.request import urlopen
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

## Getting url content

```python
# getting url html document
try:
    #source = urlopen("https://www.google.com/covid19-map/").read()
    source= urlopen("https://en.m.wikipedia.org/wiki/Template:COVID-19_pandemic_data").read()
except Exception as e:
    print(e)
```

## Converting to BeautifulSoup Object

```python
%%time
# Converting to A BeautifulSoup object
try:
    soup= BeautifulSoup(source,"lxml")
    Cases= soup.find_all('tr')
except Exception as e:
    print(e)
```

```python
%%time

try:
    soup2= BeautifulSoup(source,'lxml')
    th= soup2.find_all('th')
except Exception as e:
    print(e)
    
th=th[11:-52:2]
len(list(th))
```

## Extraction of Data

```python
# Extracting Countries
countries=[]
for i in range(0,228):
    tex = th[i].text.replace("\n",'')
    countries.append(tex)
    
countries
```

```python
countries= np.array(countries).reshape(228,1)
```

```python
df1 = pd.DataFrame(countries, columns=['Contry'])
df1
```

```python
%%time
# Extracting Value From Soup
data=[]
for row in Cases:
    td=[td.string for td in row]
    data.append(td)
cols=data[2]
array= np.array(data)

    

df3 = pd.DataFrame(array)
df3
```

```python

ndata =[]
for row in range(2,230):
    s= str(np.array(df3)[row]).split()
    ndata.append(s[5])
    ndata.append(s[7])
    ndata.append(s[9])

#removing \n
newdata=[]
for data in ndata:
    val = data.replace("\\n","").replace("'","").strip()
    newdata.append(val)
array= np.array(pd.DataFrame(newdata))
array
```

## Removing comma from 

```python

%%time
data=[]
for row in array:
    for item in row:
        if "," in item:
            data.append("".join(item.split(",")))
        elif item == " ":
            pass
        else:
            data.append(item)
nparray= np.array(data).reshape(228,3)

print(nparray)
```

## Removing Null Spaces

## Covid data to DataFrame

```python
# Convert To DataFrome
df=pd.DataFrame(nparray, columns=['Confirmed','Deaths','Recovered'])
df.replace({'None':np.nan},inplace=True)
df
# Converting object to numeric data type
df1["Confirmed"]=pd.to_numeric(df["Confirmed"])
df1["Recovered"]=pd.to_numeric(df["Recovered"])
df1["Deaths"]=pd.to_numeric(df["Deaths"])

df1
```
